# Creates a custom color palette

Let's imagine that we have a 2D array and we plotted it.

```python
import numpy as np
import matplotlib.pyplot as plt

# creates a 2D array
X = np.arange(-6, 6, 0.1)
Y = np.arange(-6, 6, 0.1)
X, Y = np.meshgrid(X, Y)
Z = X*Y*np.sqrt(X**2 + Y**2)
```

```python
# we makes the plot with matplotlib
fig = plt.figure('example', figsize=(6,4), dpi=80)
img = plt.imshow(Z)
cbar = plt.colorbar(img)
plt.show()
plt.close()
```

The plot looks fine, however using a custom palette we could highlight the regions we want. To do that we will use the **custom_color_palette** package.\
First, we import the package.

```python
import custom_color_palette as ccp
```

Then, we define the colors of the palette and make the plot.

```python
# defines color list that we will use to creates a custom palette
Colors = ['red','orange','gold','white','cyan','blue','darkblue']

# defines intervals of each color. In this case, the intervals go from -160 until 160, each 10 units.
Intervals = ccp.range(-160.0,160.0,10.0)

# defines a sublist with characteristics of colors that will be used to create a custom palette
palette1 = [Colors, Intervals]

# we pass the parm_color inside a list to the creates_palette module
cmap, ticks, norm, bounds = ccp.creates_palette([palette1])

# we makes the plot with matplotlib using the parameters gotten from creates_palette module
fig = plt.figure('example', figsize=(6,4), dpi=80)
img = plt.imshow(Z, cmap=cmap, norm=norm)
cbar = plt.colorbar(img)
plt.show()
```

In the above example, the custom palette was created using [Matplotlib's colors name](https://matplotlib.org/examples/color/named_colors.html), but also can be created using [Matplotlib's Colormap](https://matplotlib.org/tutorials/colors/colormaps.html), Hex color code or RGB color code. The following example shows an example using the combination of all of them.

```python
# defines characteristics of colors for the custom palette
palette1 = [plt.cm.autumn, ccp.range(-160.0, 0.0, 10.0)]
palette2 = [[(0/255, 255/255, 255/255),'#0080FF','blue'], ccp.range(0.0, 160.0, 10.0)]

# pass parameters to the creates_palette module
cmap, ticks, norm, bounds = ccp.creates_palette([palette1, palette2])

# creates the plot
fig = plt.figure('example', figsize=(6,4), dpi=80)
img = plt.imshow(Z, cmap=cmap, norm=norm)
cbar = plt.colorbar(img)
plt.show()
```

If you want to create a custom palette with a smaller color variation, you can do it changing the colors intervals of the palette. In previous example the intervals between colors was **10 units**, if we change it to **1 unit**, we will have the follow result:

```python
# defines characteristics of colors for the custom palette
palette1 = [plt.cm.autumn, ccp.range(-160.0, 0.0, 1.0)] # intervals each 1 unit
palette2 = [[(0/255, 255/255, 255/255),'#0080FF','blue'], ccp.range(0.0, 160.0, 1.0)] # intervals each 1 unit

# pass parameters to the creates_palette module
cmap, ticks, norm, bounds = ccp.creates_palette([palette1, palette2])

# creates the plot
fig = plt.figure('example', figsize=(6,4), dpi=80)
img = plt.imshow(Z, cmap=cmap, norm=norm)
cbar = plt.colorbar(img)
plt.show()
```

Let's go back to the previous example (where the palette had colors intervals each 10 units) and let's imagine that we want to add one color in specific region of palette. In the following example we see how to do it:

```python
# defines characteristics of colors for the custom palette
palette1 = [['red','orange','gold'], ccp.range(-160.0, -40.0, 10.0)]
palette2 = [['green'], ccp.range(-40.0, 40.0, 10.0)]
palette3 = [['#00FFFF',(0/255, 0/255, 255/255)], ccp.range(40.0, 160.0, 10.0)]

# pass parameters to the creates_palette module
cmap, ticks, norm, bounds = ccp.creates_palette([palette1, palette2, palette3])

# creates the plot
fig = plt.figure('example', figsize=(6,4), dpi=80)
img = plt.imshow(Z, cmap=cmap, norm=norm)
cbar = plt.colorbar(img)
plt.show()
```

If instead of using one color you want to use a sequence of colors, you can do it as following:

```python
# defines characteristics of colors for the custom palette
palette1 = [['red','orange','gold'], ccp.range(-160.0,-40.0,10.0)]
palette2 = [['forestgreen','lime'], ccp.range(-40.0,40.0,10.0)]
palette3 = [['#00FFFF',(0/255, 0/255, 255/255)], ccp.range(40.0,160.0,10.0)]

# pass parameters to the creates_palette module
cmap, ticks, norm, bounds = ccp.creates_palette([palette1, palette2, palette3])

# creates the plot
fig = plt.figure('example', figsize=(6,4), dpi=80)
img = plt.imshow(Z, cmap=cmap, norm=norm)
cbar = plt.colorbar(img)
plt.show()
```

Now let's imagine that we want to create a palette with their extreme colors (upper and lower color of palette). For do that we have to use the **extend** parameter into the **creates_palette** module. Also is necessary add **extend** parameter into **plt.colorbar**.

```python
# defines characteristics of colors for the custom palette
palette1 = [['red','orange','gold'], ccp.range(-160.0,-40.0,10.0)]
palette2 = [['forestgreen','lime'], ccp.range(-40.0,40.0,10.0)]
palette3 = [['#00FFFF',(0/255, 0/255, 255/255)], ccp.range(40.0,160.0,10.0)]

# pass parameters to the creates_palette module
cmap, ticks, norm, bounds = ccp.creates_palette([palette1, palette2, palette3], extend='both')

# creates the plot
fig = plt.figure('example', figsize=(6,4), dpi=80)
img = plt.imshow(Z, cmap=cmap, norm=norm)
cbar = plt.colorbar(img, extend='both') # add extend parameter into the colobar
plt.show()
```

In case we want to set a specific color for upper and lower color, we have to use **lower_color** and **upper_color** parameter into the **creates_palette** module.

```python
# defines characteristics of colors for the custom palette
palette1 = [['red','orange','gold'], ccp.range(-160.0,-40.0,10.0)]
palette2 = [['forestgreen','lime'], ccp.range(-40.0,40.0,10.0)]
palette3 = [['#00FFFF',(0/255, 0/255, 255/255)], ccp.range(40.0,160.0,10.0)]

# pass parameters to the creates_palette module
cmap, ticks, norm, bounds = ccp.creates_palette([palette1, palette2, palette3], extend='both',
                                                lower_color='white',
                                                upper_color='black')

# creates the plot
fig = plt.figure('example', figsize=(6,4), dpi=80)
img = plt.imshow(Z, cmap=cmap, norm=norm)
cbar = plt.colorbar(img, extend='both')
plt.show()
```

# Advance options

Let's imagine that we have a 2D array.

```python
import numpy as np
import matplotlib.pyplot as plt

# creates a 2D array
x = np.linspace(-7, 7, 100)
y = np.linspace(-7, 7, 100)
X, Y = np.meshgrid(x, y)
Z = np.exp(-0.02*np.sqrt((X/2)**2 + Y**2))
```

Then, we plot the data using a custom color palette.

```python
# defines characteristics of colors for the custom palette
palette = [plt.cm.gray, ccp.range(0.80,1.00,0.001)]

# pass parameters to the creates_palette module
cmap, ticks, norm, bounds = ccp.creates_palette([palette], extend='both')

# creates a ticks for colorbar
cbticks = [0.8,0.82,0.84,0.86,0.88,0.9,0.92,0.94,0.96,0.98,1.0]

# creates the plot
fig = plt.figure('example', figsize=(6,4), dpi=80)
img = plt.imshow(Z, cmap=cmap, norm=norm)
cbar = plt.colorbar(img, extend='both', ticks=cbticks)
plt.show()
```

Now, let's imagine that we want to highlight some regions with a specific color while maintaining the characteristics of the palette. In this case, if you add other palette, the grays palette will be compressed and the custom palette will have two white tones, as it see in the next example.

```python
# defines characteristics of colors for the custom palette
palette1 = [plt.cm.gray, ccp.range(0.80,0.94,0.001)]
palette2 = [['firebrick','khaki','white'], ccp.range(0.94,1.00,0.001)]

# pass parameters to the creates_palette module
cmap, ticks, norm, bounds = ccp.creates_palette([palette1, palette2], extend='both')

# creates a ticks for colorbar
cbticks = [0.8,0.82,0.84,0.86,0.88,0.9,0.92,0.94,0.96,0.98,1.0]

# creates the plot
fig = plt.figure('example', figsize=(6,4), dpi=80)
img = plt.imshow(Z, cmap=cmap, norm=norm) #'pink'
cbar = plt.colorbar(img, extend='both', ticks=cbticks)
plt.show()
```

To solved this problem, is necessary add other parameter to sublist with characteristics of colors, in this case is **palette1**.\
This parameter is a list with three elements. The first element is a values range. The second and third element define the start and end values of region that will be used into the custom palette. The next example show the custom palette with only one white tones.

```python
# defines characteristics of colors for the custom palette
palette1 = [plt.cm.gray, ccp.range(0.80,0.94,0.001), [ccp.range(0.80,1.00,0.001),0.8,0.94]]
palette2 = [['firebrick','khaki','white'], ccp.range(0.94,1.00,0.001)]

# pass parameters to the creates_palette module
cmap, ticks, norm, bounds = ccp.creates_palette([palette1, palette2], extend='both')

# creates a ticks for colorbar
cbticks = [0.8,0.82,0.84,0.86,0.88,0.9,0.92,0.94,0.96,0.98,1.0]

# creates the plot
fig = plt.figure('example', figsize=(6,4), dpi=80)
img = plt.imshow(Z, cmap=cmap, norm=norm)
cbar = plt.colorbar(img, extend='both', ticks=cbticks)
plt.show()
```

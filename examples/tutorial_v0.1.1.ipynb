```python

# Let us suppose you have this satellite image of brightnes temperature (in Kelvin)
# and you want view it with a custom color palette.

# Download Satellite Image from:
# https://github.com/joaohenry23/custom_color_palette/blob/master/examples/SatImage.bin

import numpy as np
import matplotlib.pyplot as plt

SatImg = np.fromfile('SatImage.bin', dtype='float32').reshape((800,1000))

fig = plt.figure('example01', figsize=(6,4), dpi=100)
img = plt.imshow(SatImg, cmap=plt.cm.binary)
cbar = plt.colorbar(img)
plt.show()
```

```python

# Now, let suppouse that we want to create a custom color palette,
# for that let us go to import a custom_color_palette package

import custom_color_palette as ccpl

# Then, we called the creates_palette module from package alias (ccpl)
# and pass it two parameters.
# The first parameter is a list containing a sub-list with colors names
# the second parameter is a list with the range of palette values.
# You can use the matplotlib's colors names defined
# here (https://matplotlib.org/examples/color/named_colors.html), also you can create
# your custom color palette using the Hex color code or the RGB color code as seen below.

palette, clabels, norm, bounds = ccpl.creates_palette([['gold','orange','red','blue','cyan',
                                                        '#5A6673',(0/255, 0/255, 0/255)]],
                                                      [180.0,330.0])
```

```python
# Creates a tickslabels
tickslabels = np.arange(180,340,10)

# Using mypalette, colorslabels and norm, and adding the spacing parameter in imshow
# we will create the image again.

fig = plt.figure('example01', figsize=(6,4), dpi=100)
img = plt.imshow(SatImg, cmap=palette, norm=norm)
cbar = plt.colorbar(img, spacing='proportional', ticks=tickslabels)
cbar.ax.set_yticklabels(tickslabels)
plt.show()
```

```python
# Also you can use the matplotlib's colors palette
# defined here (https://matplotlib.org/tutorials/colors/colormaps.html)

palette, clabels, norm, bounds = ccpl.creates_palette([plt.cm.jet], [180.0,330.0])

tickslabels = np.arange(180,340,10)

fig = plt.figure('example02', figsize=(6,4), dpi=100)
img = plt.imshow(SatImg, cmap=palette, norm=norm)
cbar = plt.colorbar(img, spacing='proportional', ticks=tickslabels)
cbar.ax.set_yticklabels(tickslabels)
plt.show()
```

```python

# if you would like add extend colors to the palette,
# you can do it with the parameter Extend
# Note: for Extend='both', is necessary add one color in both extreme of palette
#       if Extend='min', so is necessary add one color in the inferior extreme of palette
#       if Extend='max', so is necessary add one color in the superior extreme of palette
palette, clabels, norm, bounds = ccpl.creates_palette([plt.cm.jet], [179.0,331.0], Extend='both')

tickslabels = np.arange(180,340,10)

fig = plt.figure('example02', figsize=(6,4), dpi=100)
img = plt.imshow(SatImg, cmap=palette, norm=norm)
# also is necessary set the parameter extend='both' in the colorbar
cbar = plt.colorbar(img, spacing='proportional', ticks=tickslabels, extend='both')
cbar.ax.set_yticklabels(tickslabels)
plt.show()
```

```python
# On the other hand, if you want to merge your custom color palette with the
# Matplotlib's color palette, you can do it as follows.

palette, clabels, norm, bounds = ccpl.creates_palette([['yellow','orange','red','darkred'],
                                                       plt.cm.ocean_r],
                                                      [180.0,250.0,330.0])

tickslabels = np.arange(180,340,10)

fig = plt.figure('example03', figsize=(6,4), dpi=100)
img = plt.imshow(SatImg, cmap=palette, norm=norm)
cbar = plt.colorbar(img, spacing='proportional', ticks=tickslabels)
cbar.ax.set_yticklabels(tickslabels)
plt.show()
```

```python
# In addition to that, if you want your palette to have colors that do not change its tone,
# you can do it by adding sublists with the name of the color you want, as shown below.

palette, clabels, norm, bounds = ccpl.creates_palette([['yellow','gold','red','darkred'],
                                                       ['lime'], ['hotpink'], plt.cm.ocean_r],
                                                      [180.0,250.0,255.0,260.0,330.0])

tickslabels = np.arange(180,340,10)

fig = plt.figure('example04', figsize=(6,4), dpi=100)
img = plt.imshow(SatImg, cmap=palette, norm=norm)
cbar = plt.colorbar(img, spacing='proportional', ticks=tickslabels)
cbar.ax.set_yticklabels(tickslabels)
plt.show()
```

```python

# In the case you want to change the spacing between the colors, you can do it
# using Increase parameter. This parameter is optional and its default value is 1.0.
# The nexts lines show an example where add Increase=[1.0, 1.0, 5.0 5.0] in creates_palette
# with purpose of set the colors of each palette.
# In this example the spacing of the first colors palette is equal to 1,
# meanwhile the spacing of the last colors palette is equal to 5 (pay attention to colors bar).

palette, clabels, norm, bounds = ccpl.creates_palette([['yellow','gold','red','darkred'],
                                                       ['lime'], ['hotpink'], plt.cm.ocean_r],
                                                      [180.0,250.0,255.0,260.0,330.0],
                                                      Increase=[1.0,1.0,5.0,5.0])

tickslabels = np.arange(180,340,10)

fig = plt.figure('example05', figsize=(6,4), dpi=100)
img = plt.imshow(SatImg, cmap=palette, norm=norm)
cbar = plt.colorbar(img, spacing='proportional', ticks=tickslabels)
cbar.ax.set_yticklabels(tickslabels)
plt.show()
```

```python

# Note: In the example above, the colors interval of Lime and Hot Pink colors were also defined,
# Lime equal to 1 and Hotpink equal to 5,
# however, even so, they show the same behavior because they are colors do not change
# their tone, for which spacing no have effect over they.
```

```python
# If you want to have more control over the custom color palette, you can get it
# using EditPalette parameter.
# This parameter permit define the range of color palette to then split it,
# selecting the colors region of palette that you want to use.
```

```python
# import packages
import numpy as np
import matplotlib.pyplot as plt
import custom_color_palette as ccpl

# read file
SatImg = np.fromfile('SatImage.bin', dtype='float32').reshape((800,1000))
```

```python
# Let us suppose that we have a satellite image with black and white colors palette
# and we want to the values between 180.0 K and 240.0 K to have a custom color palette.
# We could do the following:

palette, clabels, norm, bounds = ccpl.creates_palette([['maroon','red','darkorange','#ffff00',
                                                        'forestgreen','cyan','royalblue',
                                                        (148/255, 0/255, 211/255)],
                                                       plt.cm.Greys],
                                                      [180.0,240.0,330.0])

tickslabels = np.arange(180,340,10)

fig = plt.figure('example07', figsize=(6,4), dpi=100)
img = plt.imshow(SatImg, cmap=palette, norm=norm)
cbar = plt.colorbar(img, spacing='proportional', ticks=tickslabels)
cbar.ax.set_yticklabels(tickslabels)
plt.show()
```

```python
# The result of example above is not bad, however the problem whit this is that the grey color
# palette was compacted, so we see colors with white tonalities above 240.0 K.

# To solve this problem we will add EditPalette parameter to creates_palette.

palette, clabels, norm, bounds = ccpl.creates_palette([['maroon','red','darkorange','#ffff00',
                                                        'forestgreen','cyan','royalblue',
                                                        (148/255, 0/255, 211/255)],plt.cm.Greys],
                                                      [180.0,240.0,330.0],
                                                      EditPalette=[None,[180.0,330.0,240.0,330.0]])

tickslabels = np.arange(180,340,10)

fig = plt.figure('example08', figsize=(6,4), dpi=100)
img=plt.imshow(SatImg, cmap=palette, norm=norm)
cbar = plt.colorbar(img, spacing='proportional', ticks=tickslabels)
cbar.ax.set_yticklabels(tickslabels)
plt.show()
```

```python
# Note: In the example above, we add EditPalette=[None,[180.0,330.0,240.0,330.0]].
# With the first element of this list (None) indicate that the first color palette
# will not be edited, while the second element (a sublist) indicate that
# the second color palette will be edited.
# This sub-list have 4 values, the first and second values are used to define
# the range of colors palette, the third and fourth values are used to select the colors
# that we want to use. Using this parameter we notice that the colors above 240.0 K have
# grey tonalities and not white tonalities.
```

```python
# Finally, if you wish modify the interval between colors of upper pallete you can do it adding
# the Increase parameter to the the creates_palette functions.

palette, clabels, norm, bounds = ccpl.creates_palette([['maroon', 'red', 'darkorange','#ffff00',
                                                        'forestgreen','cyan','royalblue',
                                                        (148/255, 0/255, 211/255)],plt.cm.Greys],
                                                      [180.0,240.0,330.0],
                                                      EditPalette=[None,[180.0,330.0,240.0,330.0]],
                                                      Increase=[1.0,5.0])

tickslabels = np.arange(180,340,10)

fig = plt.figure('example09', figsize=(6,4), dpi=100)
img=plt.imshow(SatImg, cmap=palette, norm=norm)
cbar = plt.colorbar(img, spacing='proportional', ticks=tickslabels)
cbar.ax.set_yticklabels(tickslabels)
plt.show()
```

```python
# or if you prefer modify the interval between colors of upper and lower palette.

palette, clabels, norm, bounds = ccpl.creates_palette([['maroon', 'red', 'darkorange','#ffff00',
                                                        'forestgreen', 'cyan','royalblue',
                                                        (148/255, 0/255, 211/255)],plt.cm.Greys],
                                                      [180.0,240.0,330.0],
                                                      EditPalette=[None,[180.0,330.0,240.0,330.0]],
                                                      Increase=[5.0,5.0])

tickslabels = np.arange(180,340,10)

fig = plt.figure('example10', figsize=(6,4), dpi=100)
img=plt.imshow(SatImg, cmap=palette, norm=norm)
cbar = plt.colorbar(img, spacing='proportional', ticks=tickslabels)
cbar.ax.set_yticklabels(tickslabels)
plt.savefig('example10.png')
plt.show()
```

```python
# also, you can use Extend='both'.
# Remember add one color in both extreme of palette and, in this case,
# also in EditPalette.
palette, clabels, norm, bounds = ccpl.creates_palette([['maroon', 'red', 'darkorange','#ffff00',
                                                        'forestgreen', 'cyan','royalblue',
                                                        (148/255, 0/255, 211/255)],plt.cm.Greys],
                                                      [175.0,240.0,335.0],
                                                      EditPalette=[None,[175.0,335.0,240.0,335.0]],
                                                      Increase=[5.0,5.0],
                                                      Extend='both')

tickslabels = np.arange(180,340,10)

fig = plt.figure('example10', figsize=(6,4), dpi=100)
img=plt.imshow(SatImg, cmap=palette, norm=norm)
# and remember set the parameter extend='both' in the colorbar
cbar = plt.colorbar(img, spacing='proportional', ticks=tickslabels, extend='both')
cbar.ax.set_yticklabels(tickslabels)
plt.savefig('example10.png')
plt.show()
```
